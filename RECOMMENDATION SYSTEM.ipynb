```python
m = np.mean(num_cust)
s = np.std(num_cust)
```

```python
stan_num_cust = num_cust.apply(lambda x: ((x-m)/s))
```

```python
##Create the recommendar system by creating a function which takes a product as a parameter and outputs the most
##similar products based on spatial distance and popularity
```

```python
import scipy as sc
from scipy import spatial
```

```python
#recommender system by customer
#the smaller the number the better
def similarity(bought,other):
    spa = sc.spatial.distance.cosine(list(df_final.loc[bought]),list(df_final.loc[other]))
    pop = stan_num_cust.loc[other]/100
    return(spa+pop)
```

```python
similarity(1846,2310)
```

```python
def rec_list(bought):
    rec = pd.DataFrame(columns = ['rec'],index = df_final.index)
    for i in rec.index:
        rec.loc[i,'rec']=similarity(bought,i)
    return(rec.sort_values(by = 'rec',ascending = True))
```
